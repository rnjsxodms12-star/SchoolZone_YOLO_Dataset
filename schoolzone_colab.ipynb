```python
# Colab 환경에 맞는 경로로 yaml 파일 덮어쓰기
yaml_content = """path: /content/SchoolZone_YOLO_Dataset
train: train/images
val: val/images
test: test/images

nc: 3
names: ['person', 'vehicle', 'road_etc']
"""

with open('/content/SchoolZone_YOLO_Dataset/data.yaml', 'w') as f:
    f.write(yaml_content)

print("✅ data.yaml 경로 업데이트 완료!")
```

```python
# YOLO 라이브러리 설치
!pip install ultralytics -q

import torch
from ultralytics import YOLO

# GPU 인식 확인 (True가 나와야 정상입니다!)
print("GPU 인식 여부:", torch.cuda.is_available())

# 모델 로드 (YOLOv11 Nano 모델)
model = YOLO('yolo11n.pt')

# 학습 시작 (Epoch 30으로 세팅)
results = model.train(
    data='/content/SchoolZone_YOLO_Dataset/data.yaml',
    epochs=30,
    imgsz=640,
    batch=32,
    device=0,
    # [수정] 프로젝트 경로를 구글 드라이브로 직접 지정!
    project='/content/drive/MyDrive/SchoolZone_Project',
    name='train_run_01',
    # [추가] 5에폭마다 드라이브에 확실히 저장 (보험용)
    save_period=5,
    # [추가] 학습이 끊겼을 때 이어서 할 수 있도록 설정
    exist_ok=True
)
```

```python
# 학습이 무사히 끝났는지 확인하고, 드라이브에 있는 최종 파일 경로 출력
import os

final_weight_path = '/content/drive/MyDrive/SchoolZone_Project/train_run_01/weights/best.pt'

if os.path.exists(final_weight_path):
    print(f"🎉 학습 완료! 최적의 모델이 다음 경로에 저장되었습니다: {final_weight_path}")
else:
    print("⚠️ 파일이 아직 생성되지 않았습니다. 학습 로그를 확인해 주세요.")
```
