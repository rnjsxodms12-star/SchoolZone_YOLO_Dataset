```python
import os
import cv2
from glob import glob
from tqdm import tqdm

def resize_images_in_folder(image_dir, target_size=640):
    image_paths = glob(os.path.join(image_dir, '*.jpg'))
    
    for img_path in tqdm(image_paths, desc=f"리사이징 중: {os.path.basename(os.path.dirname(image_dir))}"):
        img = cv2.imread(img_path)
        if img is None: continue
        
        h, w = img.shape[:2]
        # 긴 축이 640보다 크면 리사이징 진행
        if max(h, w) > target_size:
            ratio = target_size / max(h, w)
            new_size = (int(w * ratio), int(h * ratio))
            resized_img = cv2.resize(img, new_size, interpolation=cv2.INTER_AREA)
            
            # 기존 이미지 덮어쓰기 (JPEG 품질 85%로 압축률 향상)
            cv2.imwrite(img_path, resized_img, [int(cv2.IMWRITE_JPEG_QUALITY), 85])

# 데이터셋 경로 설정 (경로 확인 필수!)
base_path = r'C:\Users\user\Desktop\SchoolZone_YOLO_Dataset'

# train, val, test 폴더 순회하며 이미지 줄이기
for split in ['train', 'val', 'test']:
    folder_path = os.path.join(base_path, split, 'images')
    if os.path.exists(folder_path):
        resize_images_in_folder(folder_path)
        
print("🎉 모든 이미지 리사이징 완료!")
```
